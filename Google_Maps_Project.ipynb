```python
def text_search(business_name, location, key):
    """
    Functions to search places given a query using text_search from Palces API
    Inputs:
    -Query
    -Location
    -Radius
    -Key
    
    Outputs:
    -
    """
    
    #Endpoint
    endpoint_url = "https://maps.googleapis.com/maps/api/place/textsearch/json"
    
    #Parameters
    params = {
        'name' : name,
        'location' : location,
        'key' : key
    }
    
    #Request
    response = requests.get(endpoint_url, params = params)
    
    #Results
    results = json.loads(response.content)
    
    return results
```

```python
def nearby_places(location, radius, types, key):
    """
    Function to search places nearvy given the location and radius using nearby_places from places API
    
    Inputs:
    -
    
    Outputs:
    -
    """
    
    #Endpoint
    endpoint_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    
    #Parameters
    params = {
        'location' : location,
        'radius' : radius,
        'types' : types,
        'key' : key
    }
    
    #Request
    response = requests.get(endpoint_url, params = params)
    
    #Results
    results = json.loads(response.content)
    
    return results
```
